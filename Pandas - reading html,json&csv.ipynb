# Reading json file and convert it into data frame

```python
#reading json file
import json
import pandas as pd
import requests

url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
response = requests.get(url)
json_data = response.json()
print(json_data)
```

```python
len(json_data)
```

```python
json_data[0]['user']
```

```python
user_data = json_data[0]['user']
user_data.keys()
```

```python
pd.DataFrame(json_data).head(3)
```

```python
df = pd.DataFrame(json_data).head(3)
df[['url','repository_url','labels_url']] #in this way we can convert it into desired format
```

```python
df = pd.DataFrame(json_data).head(3)
df[['url','repository_url','user']]
```

```python
import pandas as pd

# Assuming you have already loaded your JSON data into a pandas DataFrame named 'data'

# If the 'user' key contains nested JSON data, use json_normalize to flatten it into separate columns
data_normalized = pd.json_normalize(json_data[0]['user'])

# Combine the original DataFrame with the normalized DataFrame
data_combined = pd.concat([json_data, data_normalized], axis=1)

# Drop the 'user' column, as we have already expanded its contents into separate columns
data_combined.drop('user', axis=1, inplace=True)

# Now 'data_combined' contains the JSON data with 'user' keys expanded into separate columns
print(data_combined)
```
