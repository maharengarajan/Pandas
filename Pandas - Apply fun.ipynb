```python
df = pd.DataFrame({'col1':[1,2,3,4],
             'col2':[9,8,7,6],
             'col3':'raja yoga maha lakshmi'.split()})
df
```

```python
df['col4'] = df['col1']**2
```

```python
df
```

```python
df['col5']=df['col3'].str[0]
```

```python
df
```

```python
#by using apply function
def test(a):
    return a**2

df['col4_4'] = df['col1'].apply(test)
```

```python
df
```

```python
def test1(a):
    if a == 'y':
        return 'y'
    else:
        return 'z'
    
df['col6']=df['col5'].apply(test1)
df
```

```python
#find len of str in col3
df['col7']=df['col3'].apply(len)
df
```

```python
#find len of log in col2
import math
df['col8']=df['col2'].apply(math.log10)
df
```
